```python
# Instalasi modul
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans

y.shape
```

```python
#Visualasasi titik dari data yang sudah diambil sebanyak 300 samples
X,y = make_blobs(n_samples=300, centers=1, cluster_std=0.90, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
# menghitung rata rata dari samples 300 dimuat dari bentuk grafik
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# menambahkan metode K-means untuk mencari rata rata dan menampilkan pada lingkaran merah 
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0],X[:,1])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
plt.show()
```

```python
# Instalasi modul
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
# membaca data 
data = pd.read_csv('brain_size.csv', sep=';', na_values='.')
```

```python
# menampilkan lima data ke atas 
data.head()
```

```python
# menampilkan array data dari 2 sampai 3 
X = data.iloc[:, [2,3]].values
X
```

```python
# menampilkan jumlah kolom dan table dari data yang sudah ditampilkan 
X.shape
```

```python
# memvisualisasikan data secara hirarki dan menampilkan nya dalam bentuk diagram
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

```python
# membuat label berdasarkan cluster 
model = AgglomerativeClustering(n_clusters=2, affinity='euclidean',linkage='ward')
model.fit(X)
labels = model.labels_
```

```python
# memvisualisasikan data dengan style titik 
plt.scatter(X[labels==0,0], X[labels==0,1], s=50, marker='o', color='red')
plt.scatter(X[labels==1,0], X[labels==1,1], s=50, marker='o', color='blue')

#plt.scatter(X[labels==2,0], X[labels==2,1], s=50, marker='o', color='green')
#plt.scatter(X[labels==3,0], X[labels==3,1], s=50, marker='o', color='purple')
#plt.scatter(X[labels==4,0], X[labels==4,1], s=50, marker='o', color='pink')

plt.show()
```
